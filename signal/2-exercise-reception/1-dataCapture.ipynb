```python
# Run in case of ModuleNotFoundError
# %cd "D:/gym-imu/signal/signal-utils"
# %pip install -e .

from signal_utils import IMUSampleReceiver
from signal_utils import IMUSampleTimeSeriesPlotter

RECORDING_DURATION_SECONDS = 90.0

%cd "D:/gym-imu/signal/2-exercise-reception"
EXPORTED_FILE_NAME = ".test.csv"
# EXPORTED_FILE_NAME = "./capture/pull-ups.csv"
# EXPORTED_FILE_NAME = "./capture/dips.csv"
# EXPORTED_FILE_NAME = "./capture/90-deg-push-ups.csv"
# EXPORTED_FILE_NAME = "./capture/fl-pulls.csv"

receiver = IMUSampleReceiver(RECORDING_DURATION_SECONDS)
timeSeriesPlotter = IMUSampleTimeSeriesPlotter()
```
