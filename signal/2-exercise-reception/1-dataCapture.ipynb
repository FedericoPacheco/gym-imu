# Free Data Capture

Use this notebook to record exercise (pull ups, dips, 90° push ups, front level pulls, etc.) data at the gym. Make sure to also orthogonally record myself performing them to understand everything better. 

## Setup

```python
from signal_utils import IMUSampleReceiver
from signal_utils import IMUSampleTimeSeriesPlotter

RECORDING_DURATION_SECONDS = 90.0

EXPORTED_FILE_NAME = "./capture/pull-ups.csv"
# EXPORTED_FILE_NAME = "./capture/dips.csv"
# EXPORTED_FILE_NAME = "./capture/90-deg-push-ups.csv"
# EXPORTED_FILE_NAME = "./capture/fl-pulls.csv"

receiver = IMUSampleReceiver(RECORDING_DURATION_SECONDS)
timeSeriesPlotter = IMUSampleTimeSeriesPlotter()
```

## Recording

```python
await receiver.connect()
motionData = await receiver.receive(EXPORTED_FILE_NAME)
await receiver.disconnect()
```

## Time Series Plotting

```python
timeSeriesPlotter.plot(motionData)
```
